```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('zomato.csv',encoding='latin-1')
df.head(100)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Cuisines'], axis=0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df_final=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
df_final.head()
```

```python
df_final.dtypes
```
